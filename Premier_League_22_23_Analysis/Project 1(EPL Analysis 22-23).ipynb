```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load dataset

epl_df = pd.read_csv(r'C:\Users\ACER\Documents\Python Portfolio\EPL_Project\PremierLeaguePlayerStats22_23.csv')
epl_df.head()
```

```python
# create 3 new columns

epl_df['Min'] = epl_df['Min'].astype(int)
epl_df['MP'] = epl_df['MP'].astype(int)


epl_df['MinsPerMatch'] = (epl_df['Min'] / epl_df['MP']).astype(int)
epl_df['GoalsPerMatch'] = (epl_df['Gls'] / epl_df['MP']).astype(float)
epl_df['ContributionsPerMatch'] = (epl_df['G+A'] / epl_df['MP']).astype(float)

print(epl_df.head())
```

```python
# Total Goals

Total_Goals = epl_df['Gls'].sum()
print(Total_Goals)
```

```python
# Penalty Goals

Total_Penalty_Goals = epl_df['PK'].sum()
print(Total_Penalty_Goals)
```

```python
# Penalty Attempts

Total_Penalty_Attempts = epl_df['PKatt'].sum()
print(Total_Penalty_Attempts)
```

```python
# Pie chart for penalties missed vs scored

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(13,6))
pl_not_scored = epl_df['PKatt'].sum() - Total_Penalty_Goals
data = [pl_not_scored, Total_Penalty_Goals]
labels = ['Penalties missed','Penalties scored']
colors = sns.color_palette('Set2')
plt.pie(data, labels = labels, colors = colors, autopct = '%.0f%%')
plt.show()
```

```python
# Most players from which countries

nationality = epl_df.groupby('Nation').size().sort_values(ascending = False)
nationality.head(10).plot(kind = 'bar', figsize = (12,6), color = sns.color_palette("magma"))
```

```python
# Clubs with maximum players in their squad

epl_df['Squad'].value_counts().nlargest(10).plot(kind = 'bar', color = sns.color_palette("viridis"))
```

```python
# Clubs with minimum players in their squad

epl_df['Squad'].value_counts().nsmallest(10).plot(kind = 'bar', color = sns.color_palette("viridis"))
```

```python
# Players based on age group

Under20 = epl_df[epl_df['Age'] <= 20]
age20_25 = epl_df[(epl_df['Age'] > 20) & (epl_df['Age'] <= 25)]
age25_30 = epl_df[(epl_df['Age'] > 25) & (epl_df['Age'] <= 30)]
Above30 = epl_df[epl_df['Age'] > 30]
```

```python
x = np.array([Under20['Player'].count(), age20_25['Player'].count(), age25_30['Player'].count(), Above30['Player'].count()])
mylabels = ["<=20", ">20 & <=25", ">25 & <=30", ">=30"]
plt.title('Total Players with age groups', fontsize = 20)
plt.pie(x, labels = mylabels, autopct = "%.1f%%")
plt.show()
```

```python
# Total Under 20 players in each club 

players_under_20 = epl_df[epl_df['Age'] < 20]
players_under_20['Squad'].value_counts().plot(kind = 'bar', color = sns.color_palette("cubehelix"))
```

```python
# Average age of players in each club (Ascending order)

plt.figure(figsize=(12, 6))
sns.boxplot(x='Squad', y='Age', data=epl_df, order=sorted(epl_df['Squad'].unique()))
plt.xticks(rotation=90)
```

```python
num_player = epl_df.groupby('Squad').size()
data = (epl_df.groupby('Squad')['Age'].sum())/ num_player
data.sort_values(ascending = False)
```

```python
# Total assists from each club

Assists_by_clubs = pd.DataFrame(epl_df.groupby('Squad', as_index=False)['Ast'].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x='Squad', y='Ast', data = Assists_by_clubs.sort_values(by="Ast"), palette='Set2')
ax.set_xlabel("Squad", fontsize=30)
ax.set_ylabel("Ast", fontsize=20)
plt.xticks(rotation = 75)
plt.rcParams["figure.figsize"] = (20,8)
plt.title('Plot of Clubs vs Total Assists ', fontsize = 20)
```

```python
# Top 20 Assists

top_20_assists = epl_df[['Player', 'Squad', 'Ast', 'MP']].nlargest(n=20, columns = 'Ast')
top_20_assists
```

```python
# Top 20 Goals

top_20_goals = epl_df[['Player', 'Squad', 'Gls', 'MP']].nlargest(n=20, columns = 'Gls')
top_20_goals
```

```python
# Total goals from each club

Goals_by_clubs = pd.DataFrame(epl_df.groupby('Squad', as_index=False)['Gls'].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x='Squad', y='Gls', data = Goals_by_clubs.sort_values(by="Gls"), palette='Set2')
ax.set_xlabel("Squad", fontsize=30)
ax.set_ylabel("Gls", fontsize=20)
plt.xticks(rotation = 75)
plt.rcParams["figure.figsize"] = (20,8)
plt.title('Plot of Clubs vs Total Goals ', fontsize = 20)
```

```python
# Piechart - Goals with assist and without assists

plt.figure(figsize=(14,7))
assists = epl_df['Ast'].sum()
data = [Total_Goals - assists, assists]
labels = ['Goals without assists', 'Goals with assists']
color = sns.color_palette('Set1')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```
